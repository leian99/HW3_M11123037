```python
#讀取csv
import pandas as pd

df = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")
df
```

```python
#需要做正規化的資料
from sklearn.preprocessing import LabelEncoder

le_family_history_with_overweight = LabelEncoder()
le_FAVC = LabelEncoder()
le_CAEC = LabelEncoder()
le_SMOKE = LabelEncoder()
le_SCC = LabelEncoder()
le_CALC = LabelEncoder()
le_MTRANS = LabelEncoder()
le_NObeyesdad = LabelEncoder()
```

```python
#正規化的資料會是新一筆的資料，所以有做正規化的資料欄位名稱後面會加 _n 做區別，下一行程式碼會再改回原本名稱
df['family_history_with_overweight_n'] = le_family_history_with_overweight.fit_transform(df['family_history_with_overweight'])
df['FAVC_n'] = le_FAVC.fit_transform(df['FAVC'])
df['CAEC_n'] = le_CAEC.fit_transform(df['CAEC'])
df['SMOKE_n'] = le_SMOKE.fit_transform(df['SMOKE'])
df['SCC_n'] = le_SCC.fit_transform(df['SCC'])
df['CALC_n'] = le_CALC.fit_transform(df['CALC'])
df['MTRANS_n'] = le_MTRANS.fit_transform(df['MTRANS'])
df['NObeyesdad_n'] = le_NObeyesdad.fit_transform(df['NObeyesdad'])
df
```

```python
#丟掉重複的欄位並改回原來的名稱
x_drop = df.drop(['family_history_with_overweight','FAVC','CAEC','SMOKE','SCC','CALC','MTRANS','NObeyesdad'], axis=1)
x_n = x_drop.rename(columns={'family_history_with_overweight_n': 'family_history_with_overweight','FAVC_n': 'FAVC','CAEC_n': 'CAEC','SMOKE_n': 'SMOKE','SCC_n': 'SCC','CALC_n': 'CALC','MTRANS_n': 'MTRANS','NObeyesdad_n': 'NObeyesdad'})
x_n
```

```python
#因為Gender沒有順序性，所以做另一種的正規化
pd.get_dummies(x_n['Gender'])
```

```python
#將Gender做onehot_encoding
onehot_encoding = pd.get_dummies(x_n['Gender'], prefix = 'Gender')
```

```python
#將未正規化的Gender欄位移除，並檢視
x_1 = x_n.drop(['Gender'],axis=1)
x_1
```

```python
#合併正規化後的Gender到原來的資料，並檢視
x_4 = pd.concat([onehot_encoding,x_1],axis=1)
x_4
```

```python
#因為有部分數值有到小數點第五位，根據Height必須到小數點第二位為原則，所以將所有欄位的數值有到小數點第五位的四捨五入到小數點第二位
import math
import numpy as np

rep_Age = np.round(x_4[('Age')].values,2)
rep_Height = np.round(x_4[('Height')].values,2)
rep_Weight = np.round(x_4[('Weight')].values,2)
rep_FCVC = np.round(x_4[('FCVC')].values,2)
rep_NCP = np.round(x_4[('NCP')].values,2)
rep_CH2O = np.round(x_4[('CH2O')].values,2)
rep_FAF = np.round(x_4[('FAF')].values,2)
rep_TUE = np.round(x_4[('TUE')].values,2)
```

```python
#將原來有到小數點第五位的數值都替代成小數點第二位(四捨五入)，並檢視
x_3 = x_4.replace(x_4[('Age')].values,rep_Age)
x_2 = x_3.replace(x_3[('Height')].values,rep_Height)
x_5 = x_2.replace(x_2[('Weight')].values,rep_Weight)
x_6 = x_5.replace(x_5[('FCVC')].values,rep_FCVC)
x_7 = x_6.replace(x_6[('NCP')].values,rep_NCP)
x_8 = x_7.replace(x_7[('CH2O')].values,rep_CH2O)
x_9 = x_8.replace(x_8[('FAF')].values,rep_FAF)
x_10 = x_9.replace(x_9[('TUE')].values,rep_TUE)
x_10
```

```python
#將要預測的欄位做移除，並檢視
x = x_10.drop(['NObeyesdad'], axis = 1)
x
```

```python
#沒有要丟進模型作訓練的欄位，即應該要預測成這樣的數值
y_true = x_10['NObeyesdad'].values
y_true
```

```python
#KMeans來分群
import time
from sklearn import cluster

start = time.perf_counter()

kmeans_fit = cluster.KMeans(n_clusters = 7).fit(x)

end = time.perf_counter()
```

```python
#模型訓練的時間
exetime = end - start
print("執行時間:", round(exetime,5))
```

```python
#根據模型做的預測
y_pred = kmeans_fit.labels_
y_pred
```

```python
#使用V-measure檢測KMeans分群的結果
from sklearn import metrics
V_measure_score = metrics.v_measure_score(y_true, y_pred)
print("V-measure:",V_measure_score)
```

```python
#使用純度這個指標來檢測KMeans分群的結果
def purity_score(y_true, y_pred):
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)

purity = round(purity_score(y_true,y_pred),5)
print("純度為",purity)
```

```python
# 階層式分群
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import time
from sklearn.cluster import AgglomerativeClustering

start = time.perf_counter()

hc = AgglomerativeClustering(n_clusters=7,affinity='euclidean',linkage='ward')
hc.fit_predict(x)

end = time.perf_counter()

exetime = end - start
print("執行時間:", round(exetime,5))
```

```python
y1_pred = hc.fit_predict(x)
y1_pred
```

```python
import scipy.cluster.hierarchy as sch
dis=sch.linkage(x,metric='euclidean',method='ward')
sch.dendrogram(dis)
plt.title('Hierarchical Clustering')
plt.show()
```

```python
from sklearn import metrics
V_measure_score = metrics.v_measure_score(y_true, y1_pred)
print("V-measure:",V_measure_score)
```

```python
def purity_score(y_true, y1_pred):
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y1_pred)
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)

purity = round(purity_score(y_true,y1_pred),5)
print("純度為",purity)
```
